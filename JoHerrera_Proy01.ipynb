```python

import pandas as pd
datos = { "fecha":["20240101", "20240102", "20240103", "20240104", "20240105"]
         ,"Prod":["Prod01", "Prod02", "Prod01", "Prod03", "Prod04"]
         ,"Precio":[100, 200, 100, 300, 400]
         ,"Cantidad":[5.5, 3, 2, 6, 7]
         }

dfVentas = pd.DataFrame(datos)

dfVentas
```

```python
dfVentas.sort_values("Prod") #Desordeno proproducto
dfVentas
dfVentas.sort_values("fecha") #Ordeno por fecha



dfVentas['total vtas'] = dfVentas['Precio']*dfVentas['Cantidad']
dfVentas
```

### Punto 4

```python
dfVentas['total vtas'] = dfVentas['Precio']*dfVentas['Cantidad']
dfVentas
```

```python
dfVentas['Descuento'] = dfVentas['total vtas']*(5/100)
dfVentas
```

#Ejercicio No 3
Escribir programa que genere y muestre por pantalla un DataFrame con los datos de la siguiente tabla:

```python
import pandas as pd
datosNo3 = { "MES":["Enero", "Febrero", "Marzo", "Abril", "Mayo"]
            ,"VENTAS":[30500, 35600, 28300, 33900, 42500]
            ,"GASTOS":[22000, 23450, 18100, 35700, 32450]
         }

dfVentas3 = pd.DataFrame(datosNo3)

dfVentas3
```

##1. Columna 1 = Gasto/Ventas*100.

```python
dfVentas3["Columna 1"] = dfVentas3["GASTOS"] / dfVentas3["VENTAS"] * 100
dfVentas3
```

##2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.

```python
dfVentas3.loc[dfVentas3["VENTAS"] >= 30000, "Columna 2" ] = "Meta Superada"
dfVentas3.loc[dfVentas3["VENTAS"] < 30000, "Columna 2" ] = "Meta no Superada"
dfVentas3

##df.loc[df['set_of_numbers'] <= 4, 'equal_or_lower_than_4?'] = 'True'
```

##Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.

```python
dfVentas3.loc[dfVentas3["GASTOS"] > dfVentas3["VENTAS"], "Columna 3" ] = dfVentas3["VENTAS"] / dfVentas3["GASTOS"] * 100
dfVentas3
```

##4. Obtener el total de las ventas de los últimos 3 meses

```python

dfVentas3.tail(3)

dfVentas3.loc[ 2 : 4, ["VENTAS" ]].sum()

```

##Obtener el total de gastos de febrero a abril

```python
dfVentas3.loc[ 1 : 3, ["GASTOS" ]].sum()
```
